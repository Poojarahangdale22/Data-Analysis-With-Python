## 1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the housing dataset from Excel

housing_data = pd.read_excel('housing+(1).xlsx')

# Display the first few rows of the dataset to verify the data was loaded correctly
print(housing_data.head())


median_income_1 = 'median_income'

# Calculate the average median income
average_median_income = housing_data[median_income_1].mean()


# Plot a histogram to visualize the distribution of median income

plt.figure(figsize=(8, 6))
plt.hist(housing_data[median_income_1], bins=30, edgecolor='k')
plt.title("Median Income Distribution")
plt.xlabel("Median Income")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Plot a box plot to visualize the distribution
plt.figure(figsize=(8, 6))
plt.boxplot(housing_data[median_income_1])
plt.title("Box Plot of Median Income")
plt.ylabel("Median Income")
plt.grid(True)
plt.show()

# Print the average median income
print("Average Median Income:", average_median_income)

```

**Histogram:**

The histogram represents the frequency or count of observations within specified intervals  of the median income values.
In the x-axis, you have the ranges of median income values , and on the y-axis, you have the frequency or count of occurrences in each bin.


**box plot**

The box represents the interquartile range (IQR), with the line inside the box representing the median.
The "whiskers" extend to the minimum and maximum values within a certain range.
Dots beyond the whiskers are considered potential outliers.
  
Now, regarding the average median income:

- The calculated average median income is approximately 3.87. This value pertains to the `median_income` feature, which is continuous and represents the median income of households in the dataset.

# 4. Create a data set by deleting the corresponding examples from the data set for which total_bedrooms are not available.



```python
# Create a data set by deleting the corresponding examples from the data set for which total_bedrooms are not available.
import pandas as pd

# Assuming you have loaded your housing dataset into a variable called 'housing_data'

# Create a new dataset with rows where 'total_bedrooms' is not NaN
cleaned_housing_data = housing_data.dropna(subset=['total_bedrooms'])

# Optionally, you can reset the index of the new dataset
cleaned_housing_data.reset_index(drop=True, inplace=True)

# Now 'cleaned_housing_data' contains only the rows where 'total_bedrooms' is available

print(cleaned_housing_data.head())
```

## 5. Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set

```python
import pandas as pd

# Assuming you have loaded your housing dataset into a variable called 'housing_data'

# Calculate the mean value of 'total_bedrooms' from the original dataset
mean_total_bedrooms = housing_data['total_bedrooms'].mean()

# Create a new dataset with missing values in 'total_bedrooms' filled with the mean
filled_housing_data = housing_data.fillna({'total_bedrooms': mean_total_bedrooms})

# Now 'filled_housing_data' contains the original data with missing 'total_bedrooms' values filled with the mean
print(filled_housing_data.head())
```

# 8. Create a data set for which the ocean_proximity is ‘Near ocean’.

```python
import pandas as pd

# Assuming you have loaded your housing dataset into a variable called 'housing_data'

# Create a new dataset where 'ocean_proximity' is 'Near Ocean'
near_ocean_data = housing_data[housing_data['ocean_proximity'] == 'Near Ocean']

# Optionally, you can reset the index of the new dataset
near_ocean_data.reset_index(drop=True, inplace=True)

# Now 'near_ocean_data' contains only the rows where 'ocean_proximity' is 'Near Ocean'
print(near_ocean_data.head())
```

# 9. Find the mean and median of the median income for the data set created in question 8.

```python
import pandas as pd

# Assuming you have created the 'near_ocean_data' dataset as mentioned in question 8

# Extract the 'median_income' column from the 'near_ocean_data' dataset
median_income_near_ocean = near_ocean_data['median_income']


# Calculate the mean and median of the 'median_income' column
mean_median_income = median_income_near_ocean.mean()
median_median_income = median_income_near_ocean.median()

# Print the results
print("Mean Median Income:", mean_median_income)
print("Median Median Income:", median_median_income)
```

## 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.



```python
import pandas as pd

# Assuming you have loaded your housing dataset into a variable called 'housing_data'

# Define a custom function to categorize bedroom sizes

def categorize_bedroom_size(total_bedrooms):
    if total_bedrooms <= 10:
        return 'Small'
    elif 10 < total_bedrooms < 1000:
        return 'Medium'
    else:
        return 'Large'

# Apply the custom function to create the 'total_bedroom_size' column
housing_data['total_bedroom_size'] = housing_data['total_bedrooms'].apply(categorize_bedroom_size)

# Display the updated dataset with the new column
print(housing_data.head())
```
